```python
import pandas as pd
from pathlib import Path

# Définir les chemins
PROJECT_ROOT = Path("..").resolve()
DATA_PROCESSED = PROJECT_ROOT / "data" / "processed"
DATA_OUTPUTS = PROJECT_ROOT / "data" / "outputs"

# Chargement des données nettoyées
df = pd.read_csv(DATA_PROCESSED / "processed_data.csv")

# Sélection des colonnes de score
score_columns = [col for col in df.columns if col.endswith('_score')]

# Création du DataFrame des compétences
df_competences = df[["enseignant_uid"] + score_columns].copy()

# Calcul d’un score global (moyenne des scores)
df_competences["score_global"] = df_competences[score_columns].mean(axis=1)

# Export
df_competences.to_csv(DATA_OUTPUTS / "df_competences.csv", index=False)

# Aperçu
print("✅ df_competences exporté avec", df_competences.shape[0], "enseignants et", df_competences.shape[1], "variables")
df_competences.head()
```
